## 04_Apply, Map

```python
# 56 인덱스 안가져오기 index_col=0
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
df.shape
```

```python
#57
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e'
}

df['newIncome'] = df.Income_Category.map(lambda x: dic[x])
df['newIncome']
```

```python
# 58
def changeCT(x):
    if x == 'Unkown':
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    else:
        return 'e'
df['newIcome'] = df.Income_Category.apply(changeCT)
df['newIncome']
```

```python
# 59
df['AgeState'] = df.Customer_Age.map(lambda x: x // 10 *10 )
df['AgeState'].value_counts().sort_index()
```

```python
# 60 
# def change(x):
#     if 'Graduate' in x:
#         return 1 
#     else:
#         return 0


# df['Education_Level'].apply(change).value_counts()

df['newEduLevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

```python
# 61 
df['newLimit'] = df.Credit_Limit.map(lambda x: 1 if x >= 4500 else 0)
df['newLimit'].value_counts()
```

```python
# 62
def newct(x):
    if x['Marital_Status'] == 'Married' and x['Card_Category'] == 'Platinum':
        return 1 
    else:
        return 0


df['newState'] = df.apply(newct, axis=1)
df['newState'].value_counts()
```

```python
# 63
df['Gender'] = df.Gender.map(lambda x: 'male' if x == 'M' else 'female')
df['Gender'].value_counts()
```

### 서울시 미세먼지 데이터

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df.tail()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+ hour) + datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+ hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df['(년-월-일:시)'] 
```

```python
# 77 
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df['dayName']
```

### 07_Merge, Concat

```python
import pandas as pd 

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col=0)
df.head()
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
# 91 
pd.concat([df1,df2])
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
# 92
pd.concat([df3,df4], join='inner')
```

```python
pd.concat([df3,df4], join='outer').fillna(0)
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
pd.merge(df5, df6, on='Algeria', how='inner')
```

```python
pd.merge(df5, df6, on='Algeria', how='outer')
```
